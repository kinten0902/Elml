```python
import numpy as np
import re
import pandas as pd
import pickle
from collections import Counter
```

```python
all_df = pd.read_pickle("data/all.pickle")
# all_df = all_df.drop_duplicates(subset=['body', 'name'])
all_df
```

```python
print("データの総数：", all_df.shape[0])
print("aの中に各ラベルの数：", Counter(all_df["16types_a"]))
print("bの中に各ラベルの数：", Counter(all_df["16types_b"]), "\n")

drop_16types = all_df.drop_duplicates(subset=['body', 'name'])
print("重複投稿を排除したデータの総数：", drop_16types.shape[0])
print("有効の総数（論文）：11504")
print("aの中に各ラベルの数：", Counter(drop_16types["16types_a"]))
print("bの中に各ラベルの数：", Counter(drop_16types["16types_b"]), "\n")

last_16types = drop_16types[drop_16types['16types_a'] == drop_16types[
    '16types_b']].reset_index(drop=True)
print("重複投稿を排除、一致したデータの総数：", last_16types.shape[0])
print("一致の総数（論文）：8015")
print(Counter(last_16types["16types_a"]))
```
