```python
import glob
import numpy as np
import math
from openpyxl import Workbook
from openpyxl import load_workbook
import mojimoji
import pandas as pd
from collections import Counter
```

```python
files = glob.glob("data/excel/data_new_2.xlsx")
orig_xs, orig_ts, orig_ts2 = [], [], []
total_count = 0
linenum = {}
for fn in files:
    wb = load_workbook(filename=fn, data_only=True)
    ws = wb.active
    ws_data = np.asarray([x for x in ws.columns]).T[2:]
```

```python
orig = []
total_count = 0
prev_say = ("", "")

for line in ws_data:
    s = line[11].value
    ts = [line[13].value, line[15].value]
    who = line[8].value
    say_id = line[0].value
    reply_to = line[1].value
    if type(s) is not type(None):
        total_count += 1
        if type(ts[0]) is type(None) or ts[0] == 0:
            print("A strange label is found.", total_count + 2, ts)
            prev_say = (s, who)
            continue
        if prev_say == (s, who):
            #  print("same", total_count + 2)
            prev_say = (s, who)
            continue
#         x = mojimoji.han_to_zen(str(s).replace("_x000D_", ""))

        for i in [0, 1]:
            if ts[i] in [u"受諾", u"承諾", u"同意", u"賛成", u"受託", u"受領"]:
                ts[i] = u"了承"
            if ts[i] in [u"便乗"]:
                ts[i] = u"ジョーク"
            if ts[i] in [u"はげまし"]:
                ts[i] = u"メタ"
            if ts[i] in [u"宣言"]:
                ts[i] = u"報告"
        prev_say = (s, who)

        if type(line[5].value) is int:
            group_id = line[5].value * 1234567 + line[6].value
#             if reply_to == '\\N':
#                 reply_to = -1
        else:
            print("A strange number is found.", total_count + 2)

#         orig.append((say_id, reply_to, group_id, who, x, ts[0], ts[1]))
        orig.append((say_id, reply_to, group_id, who, s, ts[0], ts[1]))
        
    else:
        print("A strange senquence is found.", total_count + 2)
orig = np.asarray(orig)
print(
    np.sum(orig[:, 5] == orig[:, 6]), "/", len(orig), "=",
    float(np.sum(orig[:, 5] == orig[:, 6])) / len(orig))
```

```python
x = Counter(orig[:, 5].tolist() + orig[:, 6].tolist())
s = 0.0
ss = sum([k for k in x.values()])
for (n, l) in sorted([(x[k], k) for k in x.keys()]):
    s += n
    print(l, "\t", n, "\t", s / ss)

print(orig.shape)
print(orig[0])
```

```python
col = [
    'say_id', 'reply_id', 'group_id', 'name', 'body', '16types_a',
    '16types_b'
]
df = pd.DataFrame(orig, columns=col)
```
